```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from ultralytics import YOLO

# Load the YOLO model
model = YOLO(r'P:\dataset\ram\runs\detect\train\weights\best.pt')  # Replace 'best.pt' with your trained model

# Class names (replace with the actual class names used during training)
class_names= ['Cool', 'Hello', 'IloveYou', 'Namaste', 'ThankYou', 'fail', 'perfect']

# Load the image using OpenCV
cap = cv2.VideoCapture(0)

# Check if the webcam is opened correctly
if not cap.isOpened():
    print("Error: Could not open video stream.")
    exit()

# Set the resolution for faster processing
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

while True:
    # Capture frame-by-frame from the webcam
    ret, frame = cap.read()
    
    if not ret:
        print("Failed to grab frame")
        break

    # Convert the frame to RGB for YOLO model (OpenCV uses BGR by default)
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    
    # Perform inference on the current frame (with a lower confidence threshold)
    results = model.predict(source=frame_rgb, save=False, conf=0.25)  # Adjust the threshold as needed
    
    # Draw the bounding boxes and class labels on the frame (specific to hand detections)
    for result in results:
        for box in result.boxes:
            # Extract the class id
            cls_id = int(box.cls[0])  # Get the class ID

            # If the detected class is related to hands or gestures
            if class_names[cls_id] in class_names:  # Replace with gesture classes
                # Extract the original bounding box coordinates
                x1, y1, x2, y2 = map(int, box.xyxy[0])

                # Adjust the height of the bounding box to focus more on hands
                box_height = y2 - y1
                box_width = x2 - x1
                
                # Shrink the bounding box height by focusing on a region around the center
                y_center = y1 + box_height // 2
                new_box_height = int(box_height * 0.5)  # 50% of the original height
                new_box_width = int(box_width * 0.8)    # 80% of the original width
                
                # Adjust the new coordinates to focus on the hand region
                new_y1 = max(y_center - new_box_height // 2, 0)
                new_y2 = min(y_center + new_box_height // 2, frame.shape[0])
                
                # Adjust x1 and x2 by the same proportion
                x_center = x1 + box_width // 2
                new_x1 = max(x_center - new_box_width // 2, 0)
                new_x2 = min(x_center + new_box_width // 2, frame.shape[1])

                # Draw the resized bounding box on the frame
                cv2.rectangle(frame, (new_x1, new_y1), (new_x2, new_y2), (0, 255, 0), 2)

                # Extract the confidence score
                conf = box.conf[0]

                # Get the class name
                class_name = class_names[cls_id]

                # Add the class name and confidence score
                label = f"{class_name}: {conf:.2f}"
                cv2.putText(frame, label, (new_x1, new_y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 0, 0), 2)

    # Display the frame with bounding boxes
    cv2.imshow('Sign Language Hand Detection', frame)

    # Break the loop if the 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
